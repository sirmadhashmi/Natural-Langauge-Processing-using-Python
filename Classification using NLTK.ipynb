```python
## do check the lenght before it should be 1000 every time this loop run it will add 1000
revlist = []
neg_rev = movie_reviews.fileids('neg')
for rev in neg_rev:
    neg_txt_rev = rev = nltk.corpus.movie_reviews.words(rev)
    review_one_str = "".join(neg_txt_rev)
    review_one_str = review_one_str.replace(',' , ',')
    review_one_str = review_one_str.replace(".",".")
    review_one_str = review_one_str.replace("\'","''")
    review_one_str = review_one_str.replace("\'","'")
    revlist.append(review_one_str)

pos_rev = movie_reviews.fileids('pos')

for rev in pos_rev:
    neg_txt_rev = rev = nltk.corpus.movie_reviews.words(rev)
    review_one_str = "".join(neg_txt_rev)
    review_one_str = review_one_str.replace(',' , ',')
    review_one_str = review_one_str.replace(".",".")
    review_one_str = review_one_str.replace("\'","''")
    review_one_str = review_one_str.replace("\'","'")
    revlist.append(review_one_str)
```
